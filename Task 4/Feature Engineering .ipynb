#  Basic feature engineering tasks:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
dataset = pd.read_csv(url)
```

```python
dataset.rename(columns = {'Sex':'Gender'},inplace = True)
```

```python
dataset.head()
```

# Create new columns from existing ones

```python
dataset['family_size'] = dataset['SibSp'] + dataset['Parch']
dataset.head()
```

#  Encode Categorical Values

```python
dataset['Gender_encoded'] = dataset['Gender'].map({'male' : 0, 'female' : 1})
dataset.head()
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

dataset['Embarked'] = dataset['Embarked'].fillna('Unknown')
dataset['Embarked_encoded'] = le.fit_transform(dataset['Embarked'])
dataset.head()
```

# Handle Missing Values

```python
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].median())
```

```python
dataset['Cabin'] = dataset['Cabin'].fillna('Unknown') 
```

```python
dataset.head()
```

#  Try correlation matrix using .corr() and seaborn.heatmap()

```python
corr_matrix = dataset.corr(numeric_only = True)

sns.heatmap(corr_matrix,annot = True, cmap = 'coolwarm')

plt.title('Titanic Correlation Matrix')
plt.show()
```

#  Normalize a column using MinMax or StandardScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Select columns to normalize
cols_to_scale = ['Age', 'Fare', 'family_size']

# Fit and transform
dataset[cols_to_scale] = scaler.fit_transform(dataset[cols_to_scale])
```

```python
# All values in 'Age', 'Fare', 'family_size' are now between 0 and 1.

dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Standardize numeric columns
dataset[cols_to_scale] = scaler.fit_transform(dataset[cols_to_scale])
```
