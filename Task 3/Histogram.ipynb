# Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
dataset = pd.read_csv(url)
```

```python
dataset.head()
```

```python
sns.histplot(data=dataset,x = 'Age', hue='Pclass',multiple='dodge', palette='deep',bins = 10,kde=True)

plt.title("Gender Count by Pclass")
plt.xlabel("Gender",fontsize = 12)
plt.ylabel("Age",fontsize = 12)
plt.tight_layout() # ensures labels and legend don't overlap.
plt.show()
```

# Explanation of terms used:

hue='Pclass'  Splits the data by Pclass, coloring each group differently.

multiple='dodge'  Displays bars for each hue group side-by-side in each bin for easy comparison.

palette='deep'  Applies a rich, pre-defined color palette for better visual distinction.

bins=10  Divides the Age range into 10 intervals (bins).

kde=True  Adds a Kernel Density Estimate curve to show the smoothed distribution.
