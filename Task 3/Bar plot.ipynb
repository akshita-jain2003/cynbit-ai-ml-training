# Bar Plot  using matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
dataset = pd.read_csv(url)
```

```python
dataset.head()
```

```python
#If you use gender_count.plot(kind='bar'), Pandas and Matplotlib handle the index and values automatically.
# plt.barh() – Horizontal Bar Chart

gender_count = dataset['Gender'].value_counts()

plt.bar(gender_count.index,gender_count.values,color = ['blue','red'],alpha = 0.7,label = ['Male','Female'],
        edgecolor = 'black',width = 0.6,linewidth = 1)

plt.xlabel('Gender',fontsize = 12)
plt.ylabel('Count',fontsize = 12)
plt.title('Number of Passengers by Gender',fontsize = 14)
plt.xticks(fontsize = 10,rotation = 0) # Customize Axis Tick Labels
plt.yticks(fontsize = 10)
plt.grid(axis='y', linestyle='--', alpha=0.5) # Add Grid Lines
plt.legend() # 
plt.show() 
```

# Bar Plot  using matplotlib.pyplot and seaborn 

```python
import seaborn as sns

sns.countplot(data=dataset, x='Gender', hue='Pclass', palette='deep',alpha = 1)

plt.title("Gender Count by Pclass")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

# Explanation of terms used :

plt.bar(...) – Creates a vertical bar chart.

gender_count.index – Categories (x-axis labels: 'Male', 'Female').

gender_count.values – Counts (heights of bars).

color=['blue', 'red'] – Sets bar colors for each category.

alpha=0.7 – Sets transparency of bars (0 = fully transparent, 1 = opaque).

label=['Male', 'Female'] – Labels for legend (note: here this will not work as expected; it adds one label for both bars).

edgecolor='black' – Adds black borders around each bar.

width=0.6 – Sets the width of the bars (default is 0.8).

linewidth=1 – Sets the thickness of the bar borders.

sns.countplot() – Creates a bar plot showing the count of observations in each category.

palette='deep' – Applies Seaborn’s default deep color palette to distinguish groups.
